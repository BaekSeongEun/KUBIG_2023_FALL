# 데이터 로딩

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#보스턴 주택 데이터셋
from sklearn import datasets
diabetes = datasets.load_diabetes()
```

```python
# 데이터 정보 확인
print(diabetes.DESCR)
```

```python
# DataFrame 으로 변환
df = pd.DataFrame(diabetes['data'], columns = diabetes['feature_names'])
target = pd.DataFrame(diabetes['target'], columns = ['target'])

print(df.shape)
df.head()
```

```python
print(target.shape)
target.head()
```

```python
# 데이터 셋 통합
df = pd.concat([df, target], axis = 1)
df.head()
```

## 추가 작업

```python
df.sex.value_counts()
```

```python
df.loc[:,'sex'] = np.where(df.loc[:,'sex'] > 0, 1, 0)
df.head()
```

```python
df.head()
```

# 데이터 전처리

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_scaled = df.iloc[:,:-1]
scaler.fit(df_scaled)
df_scaled = scaler.transform(df_scaled)
```

```python
df.iloc[:,:-1] = df_scaled[:,:]
df.head()
```

## train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size=0.2)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)

```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# fitting
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
train_pred = lr.predict(X_train)
y_pred = lr.predict(X_test)
print('train mse :', mean_squared_error(y_train, train_pred))
print('test mse :', mean_squared_error(y_test, y_pred))
```

```python
#예측값과 실제값의 분포
plt.scatter(X_test['bmi'], y_test, label='y_test')
plt.scatter(X_test['bmi'], y_pred, c='r', label='y_pred')
plt.show()
```

# Q. 각자 회귀 모델에 대해서 조사해서 본 데이터 셋으로 돌리고, 실제값과 예측값에 대해 비교해보세요!

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)

train_pred = model.predict(X_train)
y_pred = model.predict(X_test)
print('train mse :', mean_squared_error(y_train, train_pred))
print('test mse :', mean_squared_error(y_test, y_pred))


plt.scatter(X_test[:,2], y_test, label='y_test')
plt.scatter(X_test[:,2], y_pred, c='r', label='y_pred')
plt.show()
```
