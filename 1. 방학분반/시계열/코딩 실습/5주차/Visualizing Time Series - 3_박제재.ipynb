# Partial autocorrelation

* 시계열과 지연된 자신의 카피와 상관관계를 측정하는데, 일반 autocorrelation과 다르게 이전 시점의 영향도 제거.


plot_pacf 편자기상관

편 자기상관이 0에 가까우면 서로 상관되지 않는다.
절댓값이 1에 가까울 수록 지연된 관측치 간 강한 상관 양 또는 음의 상관 관계가 존재.

위 그래프에서 파란색으로 표시된 불확실성 한계 마진 그 바깥 영역에서 관찰된 부분은 편 자기상관 값이 통계적으로 유의하다고 할 수 있음.

# 시계열 decomposition
추세 or 계절성은 수동으로 식별해야 함. -> 시계열 분해 방법으로 시계열 데이터의 구조를 자동으로 추출 및 수량화 가능.

Statsmodels.tsa 모듈을 활용.
sm.tsa.seasonal_decompose! 이 함수는 (계절성, 추세, 노이즈) 를 포함하는 객체를 반환.




각 개별 구성요소를 추출 및 플롯 가능.
dir()를 통해 관련 속성 파악 가능.
이를 통해 계절 성분을 나타내고 싶으면 decomposition.seasonal 을 사용.



계절적 패턴은 계절적 요인이 있을 때 존재.


이번엔 trend, 추세를 추출해보자. 시계열 전반적 진행상황을 반영. 





마지막으로 노이즈 or 시계열 잔여 성분 추출. 추세나 계절성으로 알 수 없는 무작위적 불규칙한 영향을 설명.



```python
# 퀴즈 2-1

# Import statsmodels.api as sm
import statsmodels.api as sm

# Perform time series decompositon
decomposition = sm.tsa.seasonal_decompose(co2_levels)

# Print the seasonality component
print(decomposition.seasonal)



# <Output>
# datestamp
#     1958-03-29    1.028
#     1958-04-05    1.235
#     1958-04-12    1.412
#     1958-04-19    1.701
#     1958-04-26    1.951
#                   ...  
#     2001-12-01   -0.525
#     2001-12-08   -0.393
#     2001-12-15   -0.135
#     2001-12-22    0.116
#     2001-12-29    0.285
#     Name: seasonal, Length: 2284, dtype: float64
```

```python
#퀴즈 2-2

# Extract the trend component
trend = decomposition.trend

# Plot the values of the trend
ax = trend.plot(figsize=(12, 6), fontsize=6)

# Specify axis labels
ax.set_xlabel('Date', fontsize=10)
ax.set_title('Seasonal component the CO2 time-series', fontsize=10)
plt.show()
```
