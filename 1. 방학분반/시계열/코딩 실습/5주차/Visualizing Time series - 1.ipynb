# slicing time series data

* discoveries['1960' : '1970' ] 1960 1970 사이 데이터.
* discoveries['1950-01' : '1950-12'] 1950년 1월부터 12월까지.
* [1960 - 01 - 01 : 1960 - 01 - 15] 는 1960년 1월 1일부터 1960년 1월 15일까지 데이터 추출.



markers = 주석 : axvline axvhline을 사용.

axvline는 vertical로 1969 01 01위치에 나오게되며, axhline은 horizontal로 y축의 값이 4인 곳에서 만들어짐.

axvspan : 플롯 특정 영역 음영처리
axhspan : 수징영역과 수평영역생성 
둘다 관심영역을 강조 표시하는데 사용.


```python
# 퀴즈 5

# Select the subset of data between 1945 and 1950
discoveries_subset1 = discoveries['1945':'1950']

# Plot the time series in your DataFrame as a blue area chart
ax = discoveries_subset1.plot(color='blue')

# Show plot
plt.show()
```

```python
# 퀴즈 6

# Plot your the discoveries time series
ax = discoveries.plot(color='blue')

# Add a red vertical line
ax.axvline('1939-01-01', color='red', linestyle='--')

# Add a green horizontal line
ax.axhline(4, color='green', linestyle='--')

plt.show()
```

```python
# 퀴즈 7

# Plot your the discoveries time series
ax = discoveries.plot(color='blue', fontsize=6)

# Add a vertical red shaded region
ax.axvspan('1900-01-01', '1915', color='red', alpha = 0.3)

# Add a horizontal green shaded region
ax.axhspan(6, 8, color='green', alpha=0.3)

plt.show()
```
