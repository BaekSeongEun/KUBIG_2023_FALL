```python
df['DATE']=pd.to_datetime(df['DATE'])
```

```python
# matplotlib
import matplotlib.pyplot as plt
```

```python
# 날짜를 index로 사용
df=df.set_index('DATE')
df.plot()
plt.show()
```

```python
# plot style 변경
plt.style.use('fivethirtyeight')
df.plot()
plt.show()
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
df.plot()
plt.show()
```

```python
# 그래프 label 추가
ax=df.plot(color='green')
ax.set_xlabel('Date')
ax.set_ylabel('power value')
ax.set_title('The Power')
plt.show()
```

```python
ax=df.plot(figsize=(8,3),linewidth=2,fontsize=6, color='blue')
ax.set_xlabel('Date')
ax.set_ylabel('power value')
ax.set_title('The Power', fontsize=9)
plt.show()
```

```python
df['2000':'2010']
```

```python
# 새 데이터프레임 생성
df_subset=df['2000':'2010']
```

```python
# 추가 주석 달기 -> 수직선 수평선 그리기
ax=df_subset.plot(color='blue')
ax.set_xlabel('Date')
ax.set_ylabel('power value')
ax.set_title('The Power 2000~2010')
ax.axvline(x='2001-01-01', color='red', linestyle='--')
ax.axhline(y=100, color='green', linestyle='--')
plt.show()
```

```python
# 하이라이 강조
ax=df_subset.plot(color='blue')
ax.set_xlabel('Date')
ax.set_ylabel('power value')
ax.set_title('The Power 2000~2010')
ax.axvline(x='2001-01-01', color='red', linestyle='--')
ax.axhline(y=100, color='green', linestyle='--')
ax.axvspan('2003-01-01','2004-12-01', color='red', alpha=0.5)
ax.axhspan(90,110,color='green', alpha=0.2)
plt.show()
```

```python
ax=df_subset.plot(color='blue')
ax.set_xlabel('Date')
ax.set_ylabel('power value')
ax.set_title('The Power 2000~2010')
ax.axvline(x='2001-01-01', color='red', linestyle='--')
ax.axhline(y=100, color='green', linestyle='--')
ax.axvspan('2003-01-01','2004-12-01', color='red', alpha=0.8)
ax.axhspan(90,110,color='green', alpha=0.1)
plt.show()
```
